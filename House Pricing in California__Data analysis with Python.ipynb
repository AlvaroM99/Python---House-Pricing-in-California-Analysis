# 3.- Sale price data analysis

First things first, let's overview the SalePrice column using the .describe( ) method.

```python
df_train['SalePrice'].describe()
```

Reviewing the mean, minimum and maximum parameters, there's no apparent problem with the values. Moreover, the percentile values increase by the hand of the increasing percentile percentage, another sign of healthy values. Now let's plot the SalePrice distribution using the seaborn command **distplot**.

```python
sns.displot(df_train['SalePrice'], color='#F42272', kde=True);
```

The graph depicts the following:
- The distribution deviates from the normal distribution as it shoud be more centered and flared.
- It has a positive skewness (obliquity). The right tail is way longer that the one on the left, thus the distribution is leaning to the left and its value will be positive when analyzed with the .skew( ) method.
- The plot shows strong peakedness (kurtosis). It means that there are many houses in a very restricted price range, which causes the height of the bars to increase. The strong sharpness makes the distribution look similar to the Leptokurtic distribution.

We can find out the actual values of skewness and kurtosis of the graph using the **skew()** and **kurt()** methods.

```python
print('Skewness: ' + str(df_train['SalePrice'].skew()))
```

This positive value of skewness indicates what we saw at first sight, the distribution is shifted from the center, thus leaning to the left.

```python
print('Kurtosis: ' + str(df_train['SalePrice'].kurt()))
```

The skewness in a normal distribution is around 3, when the value is surpassed it means increasing peakedness. A flattened distribution has a lower value than 3.

### 3.1.- Sale Price vs Living Area

Let's see what's the behaviour of the "SalePrice" variable when depending of the variable "GrLivArea". We will graph a **scatter plot** using pandas to confront the sale price and the living area of the house. The aim is to figure out whether or not there is a linear relationship between the two parameters.

To visualize data using pandas we must accomplish three steps. First let's declare a new variable that contains the reference to the living area column, that is the header 'GrLivArea'. 

```python
var = 'GrLivArea'
```

The second step will be extracting the two columns of interest from the dataframe and concatenating them using the pandas method pd.concat( ) **.concat([df_name[column1_name], df_name[column2_name]], axis = 1 or 0)** to generate a unique sub-table ready to be plotted. The axis term of the method stands for the concatenation route; if the axis equals 1 then the idex are the columns, if it equals 0 the index will be the rows. (axis = 1 --> .concat( ) all the rows of the mentioned columns // axis = 0 --> .concat( ) all the columns of the mentioned rows)

```python
data = pd.concat([df_train['SalePrice'],df_train[var]], axis=1)
```

```python
data.head(10)
```

I have used the .head(n) method to make sure that the concatenation worked corectly. The third and final step is graphing the scatter plot using the following line of code **data.plot.scatter(x=var_indep, y=var_depend, ylim=(y_min,y_max))**. It is mandatory to declare the x and y variables and the y variable limits. 

```python
data.plot.scatter(x=var, y='SalePrice', ylim=(0,800000), color='#0E71FB');
```

It can be clearly seen that there is a linear relationship between the living area and the house price. The larger the area of the house, the more expensive the property will be.

### 3.3.- Sale Price vs Overall Quality

Diving into the construction variables we realize these variables are qualitative, thus the relationship with our key sale price variable will be a bit harder to graph. These qualitative variables are also called categorical features. Starting with the overall quality, lets visualize the relationship using a **box plot** from seaborn library; but first we must procede with the concatenation method to extract only the desired columns from the dataset, that is 'SalePrice' and 'OverallQual'.

```python
var = 'OverallQual'
data = pd.concat([df_train['SalePrice'],df_train[var]], axis=1)
data.head(10)
```

Now comes the difficult part. We declare two new variables **f and ax** in wich the result of **plt.subplots** will be stored. Inside the .subplots instruction we define the size of the figure (area) in which the seaborn graph will be pasted. Hence, f and ax store the figsize and the plot itself. As we are analyzing a categorical feature realitonship with the sale price, then we will make use of the box plot by typing **sns.boxplot**.

Finally, we define fig as a variable where we store the sns.boxplot graph and we will modify the y axis applying the seaborn method **.axis( )** to the mentioned fig variable. That's because the default values for the y axis are not appropiate.

```python
f, ax = plt.subplots(figsize=(8,6))
fig = sns.boxplot(x=var, y='SalePrice', data = data, palette='Greens', saturation=1)
fig.axis(ymin=0, ymax=800000);
```

We can clearly see the linearity between the quality of the house and its sale price. It's also interesting to point out how as the quality of the house increases, the boxes' width increase. This implies that the higher the quality of the house there is also a greater variability in its sale price.

### 3.4.- Sale Price vs Year of Construction

Lets do the same with the other categorical feature, the year of construction.

```python
var = 'YearBuilt'
data = pd.concat([df_train['SalePrice'], df_train[var]], axis=1)
data.head(10)
```

Since the **x-axis ticks** are dates and there are many dates, if we don't rotate them the values would clutter up the x-axis making it unreadable. For this reason we will apply at the end of the statement a new line that allows the rotation of these ticks **plt.xsticks(rotation=90)**

```python
f, ax = plt.subplots(figsize=(20,8))
fig = sns.boxplot(x=var, y='SalePrice', data=data)
fig.axis(ymin=0, ymax=800000);
plt.xticks(rotation=90);
```

We see how there is an upward trend for the price of housing when it is newer, however the dependency relationship between the sale price and the year of construction is not as critical as with the previous variables. The slope of the linear relationship proves this affirmation as it isn't as steep as with the other variables. The slight but constant linear increase could be due to the also constant increasing inflation.

It is curious to observe how the price of housing increased from the 2000s until it reached an exorbitant maximum in 2008 due to the brick bubble; and as of the year following the outbreak of said crisis, the price of housing plummeted. This behaviour can be also seen in other years, coinciding always with housing or finantial crisis, for example: 1907-1908 (global), 1925-1933 (US), 1970 (Global but less impact US), 1981 (Global) and 1990 (Global). The comparatives between the house pricing collapsing and the start of a crisis where made using the following references: https://www.scielo.cl/scielo.php?script=sci_arttext&pid=S0250-71612013000300003 and https://seekingalpha.com/article/4325524-housing-starts-are-truly-impressive

# 4.- Proving our intuition and searching new relationships

I just analysed four variables guided just by my subjective intuition, but there are many other that we should analyse. ### 4.1.- Correlation matrix

Firstly, we will put together a correlation matrix (heatmap design) where we can observe the correlation of every variable with the others and thus see which are the most correlated variables with the SalePrice variable.

To do this we are declaring a variable **corrmat = df_train.corr()** with all the numerical correlations between variables. For the time being lets obviate the warning.

```python
corrmat = df_train.corr()
```

As we did before, we define the area in which the plot will be placed by declaring f and ax variables. After this we can build the graph with seaborn applying **sns.heatmap(corrmat, vmax=1, square=True)**. The first term "corrmat" is the reference from where seaborn will extract the data (it was defined in the previous line), the second term "vmax=1" is the maximum correlation between variables that cannot be greater than 1 (this would a perfect correlation) and the last term "square=True" is going to be the shape given to the heatmap.

The vmax value selected this time is going to be 0.8 because none of the variables will have a better correlation than 0.8 with the SalePrice variable (key parameter). If we had set 1 as the value of vmax then the colors chosen by seaborn wouldn't have allowed us to clearly see the differences in the correlations between the 0.5 and 0.8 range.

```python
f, ax = plt.subplots(figsize=(12,9))
sns.heatmap(corrmat, vmax=0.9, square=True);
```

Therefore we see that the less color, the greater the correlation between the variables. The correlations that should be pointed out are those that are off the diagonal, since obviously a variable with itself has a perfect correlation. Following this, we see that for the variable of interest 'SalePrice' the strongest correlations are with OverallQual, GrLivArea and to a lesser extent (the square is darker) the variables TotalBsmtSF and YearBuilt. However, this is still subjective, since said conclusion is based on a personal appreciation of the colors and the differences in tone between them. Therefore, let us seek objectivity through numbers by zooming this heatmap graph.

### 4.2.- Numerical Correlation Matrix

In order to base our conclusions on purely objective data, lets follow the same steps but adding the value of the best ten 'SalePrice' correlated variables.

```python
k=10
```

Then we declare a new variable **cols** where these best ten correlated variables will be stored as the reference of each column, that is the header name. To achieve this we apply the **nlargest( )** method to the previously defines corrmat variable (this was the variable used to graph the heatmap plot). Inside the method we declared the number of parameters and the variable for which we want to return the best correlated parameters. The satetment used is the following: **cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index**

The **.index** method used at the end of the line allows us to put together all the retrieved columns' reference with the 'SalePrice' column header.

```python
cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index
```

```python
cols
```

These are the most correlated variables with our key sale price column.

The next step requieres declaring another new variables **cm** that will store the values for the correlations of the variables with the 'SalePrice' column and between them. To do this we will use the numpy library (np) and one of its commands **corrcoef**, that builds an array with the values of the correlations referenced in the vairable **cols**. Inside the command **corrcoef** we type in the references, that is the newly declared variable **cols**, and we apply the **.values** pandas method to print the numerical values obtained by numpy using the precedent references (df_train[cols]) and the method **.T** (.transpose( ) method) that allow us to permute the dimensions of a given array. The whole statement would be: **cm = np.corrcoef(df_train[cols].values.T)**

```python
cm = np.corrcoef(df_train[cols].values.T)
```

```python
cm
```

.
.
.

Once we have our zoomed array built the only thing left is to plot the zoomed heatmap. First we set the font size of the numbers that will be printed in the heatmap. Then we plot the zoomed heatmap using the following line: **hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)**. 

Lets explain the parameters inside this statement:
- **cm**:  Stands for the array we've just built from which seaborn will take in the data to build the heatmap.
- **cbar**:  If True, seaborn prints a color bar to the right side of the plot.
- **annot**:  If True, seaborn writes the data value in each cell. 
- **square**:  If True, seaborn plots the graph with a squared shape.
- **fmt**:  String formatting code to use when adding annotations.
- **annot_kws**:  Keyword arguments for matplotlib.axes.Axes.text() command when annot is True.
- **y/xticklabels**:  If True, prints the column names of the dataframe as labels. 

```python
sns.set(font_scale=1)
hm = sns.heatmap(cm, 
                 cbar=True, 
                 annot=True, 
                 square=True, 
                 fmt='.2f', 
                 annot_kws={'size':10}, 
                 yticklabels=cols.values, xticklabels=cols.values);
```

Observing the returned heatmap we can objectively conlude that:
- Overall quality, Housing area and Basement surface are strongly correlated with SalePrice.
- The garage variables are highly correlated with each other and with the SalePrice key variable, thus a further analysis might be interesting. Maybe this two garage variables wouldn't be that important if the dataset was obtained from any other country that wasn't the US, as it is well-known how important are cars for the americans' daily life due to the characteristic urban planning of its cities. 
- Finally, the variable YearBuilt is slightly correlated with SalePrice. This was our preliminar conclusion when we analyzed it, however we should do a little bit of time-series analysis to get this analysis completely right.

### 4.3.- Scatter plots between 'SalePrice' and correlated variables

Lastly I will graph Scatter plots of the most correlated variables. This may be redundant, but it provides lots of new insights in a very limited space, hence it's worth and cool trying.

First we set our scatter plots configuration by typing **sns.set( )**. As the parenthesis won't be filled the configuration is going to be the default one.

```python
sns.set()
```

Then we defined cols as the variable storing the references of the most correlated parameters returned in the last heatmap.

```python
cols = ['SalePrice', 'OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'FullBath', 'YearBuilt']
```

Finally we graph the plot using **sns.pairplot(df_train[cols], height=2.5)**. The **pairplot( )** by default will create a grid of axes such that each numeric variable in data will by shared across the y-axes across a single row and the x-axes across a single column. The diagonal plots are treated differently; a univariate distribution plot is drawn to show the marginal distribution of the data in each column. However, we can ignore the graphs on the diagonal. The only modification will be the **height** which will be set to 2.5 inches for each facet.

The second line **plt.show( )** is just a format thing, to vanish the header created by default for the plot.

```python
sns.pairplot(df_train[cols], height=2.5)
plt.show()
```

Although we already know some of the main figures, this mega scatter plot gives us new insights about the variables relationships between them and with the 'SalePrice'.

The figure between 'TotalBsmtSF' and 'GrLiveArea' unveils a new obvious but interesting insight. In the graph we can see some dots drawing a prefect straight line, which almost acts like a border. It makes sense that the majority of the dots stay below that line, as the basement area can be equal or smaller than the above living area surface, but it is very rare to find a basement bigger than the actual living area.

The scatter plot concerning 'SalePrice' and 'YearBuilt' also reveals more insights that were impossible to see in the box plot graphed before. At the bottom of the 'dots cloud' we see what appears to be a shy exponential function. This tendency can also be seen in the upper limit of the 'dots cloud'. These insights might be related with the fact that very old houses have the attraction of antiquities and history, on the other hand newer houses built with fresh designs and better materials are significantly more expensive. In fact the upper tendency increases faster than the bottom one, probably due to an increasingly saturated real estate market.
